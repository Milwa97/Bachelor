```python
def clean(df, n_of_samples = 500):
    df.dropna(axis = 1, thresh = n_of_samples, inplace = True)
    df.dropna(axis = 0, inplace = True)
    return df
```

```python
def show(df_X,df_Y, df_ZX, df_ZY, df_P, n_links, lambd ):
    
    df_X = clean(df_X)
    df_Y = clean(df_Y)
    df_ZX = clean(df_ZX)
    df_ZY = clean(df_ZY)
    df_P = clean(df_P)
    
    df_X_mean = df_X.mean()
    df_X_std = df_X.std()
    df_Y_mean = df_Y.mean()
    df_Y_std = df_Y.std()
    df_Z_mean = df_ZX.mean()
    df_Z_std = df_ZX.std()
    df_P_mean = df_P.mean()
    df_P_std = df_P.std()
    
    time = np.arange(len(df_P_mean))

    fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (15,10), sharex=False)

    for row in axs:
        for ax in row:
            ax.set_xlabel('time')
            ax.set_ylabel('number of cells')
            ax.xaxis.label.set_size(16)
            ax.yaxis.label.set_size(16)
            ax.grid(True)
            ax.tick_params(grid_color='gray', grid_alpha=0.5, labelcolor = "black",
                           labelsize = 15)      

    axs[0][0].set_title('cancer cells', fontsize = 18)
    axs[0][1].set_title('cytotoxic cells', fontsize = 18)
    axs[1][0].set_title('Z-complex cells', fontsize = 18)
    axs[1][1].set_title('dead cells', fontsize = 18)

    axs[0][0].errorbar(x = time, y = df_X_mean, yerr = df_X_std, color = 'red', alpha = 0.25)
    axs[0][1].errorbar(x = time, y = df_Y_mean, yerr = df_Y_std, color = 'yellow', alpha = 0.25)
    axs[1][0].errorbar(x = time, y = df_Z_mean, yerr = df_Z_std, color = 'purple', alpha = 0.25)
    axs[1][1].errorbar(x = time, y = df_P_mean, yerr = df_P_std, color = 'black', alpha = 0.25)

    axs[0][0].plot(time, df_X_mean, color = 'red')
    axs[0][1].plot(time, df_Y_mean, color = 'yellow')
    axs[1][0].plot(time, df_Z_mean, color = 'purple')
    axs[1][1].plot(time, df_P_mean, color = 'black')

    fig.suptitle('lambda = {:}, n_links = {:}'.format(lambd, n_links))
    
    fig.tight_layout()      
    plt.show()
```
