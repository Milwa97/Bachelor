```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
    
sns.set(style="darkgrid")

NAMES = [5,10,15,20,25,30,35, 45, 50]
```

```python
def read_data(filename, skip_rows = 4, treshold = 4):
    df = pd.read_csv( filename, skiprows = skip_rows, names=NAMES)
    df = df.dropna(axis = 'columns', how = 'all')
    df = df.dropna(axis = 'index', thresh=3)
    df = df.fillna(1500)
    return df
```

```python
df175 = read_data('c175X25.csv')
df150 = read_data('c15X25.csv')
```

```python
def histogram(df):

    f, axes = plt.subplots(1, 3, figsize=(15, 5))
    sns.despine(left=True)

    sns.distplot(df[10], color="m", axlabel = "range 10", bins = 40, label = 'range 10', ax=axes[0], kde = False)
    sns.distplot(df[15], color="g", axlabel = "range 15", bins = 40, label = 'range 15', ax=axes[1], kde = False)
    sns.distplot(df[20], color="b", axlabel = "range 20", bins = 40, label = 'range 20', ax=axes[2], kde = False)

    plt.tight_layout()
```

```python
histogram(df175)
```

```python
histogram(df150)
```
