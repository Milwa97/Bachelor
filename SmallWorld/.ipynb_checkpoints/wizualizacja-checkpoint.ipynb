```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
    
sns.set(style="darkgrid")

m_range = [0,5,10,20,30,40,50,60]
```

```python
def read_data (filename, skip_rows = 9, treshold = 6):
    df = pd.read_csv( filename, skiprows = skip_rows, names= m_range)
    
    for i in range(len(df[10]) ):
        if (df[10][i] == df[5][i]):
            df[20][i] = df[10][i]
    
    for m in [20,30,40,50,60]:
        for i in range( len(df[10])):
            if (df[m][i] == df[m-10][i] and m < 60):
                df[m+10][i] = df[m][i]                

    df.dropna(axis = 'index', thresh = treshold, inplace = True)
    
    return df.astype(dtype = 'int32')
```

```python
def histogram(df):

    f, axes = plt.subplots(1, 4, figsize=(15, 5))
    sns.despine(left=True)

    sns.distplot(df[30], color="r", axlabel = "range 30", bins = 20, label = 'range 30', ax=axes[0], kde = False)
    sns.distplot(df[40], color="m", axlabel = "range 40", bins = 20, label = 'range 40', ax=axes[1], kde = False)
    sns.distplot(df[50], color="b", axlabel = "range 50", bins = 20, label = 'range 50', ax=axes[2], kde = False)
    sns.distplot(df[60], color="g", axlabel = "range 60", bins = 20, label = 'range 60', ax=axes[3], kde = False)

    plt.tight_layout()
```

```python
df_n100c10X100l03 = read_data('n100c10X100l03.csv')
df_n100c10X100l1 = read_data('n100c10X100l1.csv')
df_n100c15X100l03 = read_data('n100c15X100l03.csv')
df_n100c15X100l1 = read_data('n100c15X100l1.csv')
df_n100c20X100l03 = read_data('n100c20X100l03.csv')
df_n100c20X100l1 = read_data('n100c20X100l1.csv')
```
