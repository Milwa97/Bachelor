```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
s = np.arange(0, 50, 0.01)

KM = [1,1,5,5]
V = [2,3,2,3]

y0 = V[0] *s / (KM[0] +s)
y1 = V[1] *s / (KM[1] +s)
y2 = V[2] *s / (KM[2] +s)
y3 = V[3] *s / (KM[3] +s)


fig, ax = plt.subplots(figsize = (15,10) )

        
ax.yaxis.label.set_size(18)
ax.xaxis.label.set_size(18)
    
ax.set_ylabel('reaction rate')
ax.set_xlabel('substrate concentration')
ax.set_xlim([0,50])
ax.set_ylim([0,3.5])

ax.tick_params(grid_color='gray', grid_alpha=0.5, labelcolor = "black", labelsize = 15)      
ax.grid(False)  
 

ax.plot(s,y0, c = 'red', label = "K = {0:}  Vmax = {1:}".format(KM[0], V[0]) )
ax.plot(s,y1, c = 'magenta', label = "K = {0:}  Vmax = {1:}".format(KM[3], V[1]) )
ax.plot(s,y2, c = 'blue', label = "K = {0:}  Vmax = {1:}".format(KM[2], V[2]) )
ax.plot(s,y3, c = 'green', label = "K = {0:}  Vmax = {1:}".format(KM[1], V[3]) )
    
ax.legend( loc="lower right", fontsize = 15, title="Parameters", title_fontsize = 16)

ax.hlines(y = V[1], xmin = -5, xmax = 55, color = 'gray', linestyle = '--')
ax.hlines(y = V[2], xmin = -5, xmax = 55, color = 'gray', linestyle = '--')

ax.vlines(x = KM[0], ymin = -5, ymax = 55, color = 'red', linestyle = '--', alpha = 0.2)
ax.vlines(x = KM[1], ymin = -5, ymax = 55, color = 'magenta', linestyle = '--', alpha = 0.2)
ax.vlines(x = KM[2], ymin = -5, ymax = 55, color = 'blue', linestyle = '--', alpha = 0.2)
ax.vlines(x = KM[3], ymin = -5, ymax = 55, color = 'green', linestyle = '--', alpha = 0.2)

#fig.savefig(filename)
```
